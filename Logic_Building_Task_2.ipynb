```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
present = attendance.count("P")
percentage = (present / len(attendance)) * 100
print("Attendance Percentage:", percentage)
```

```python
text = "python"
frequency = {}
for char in text:
    frequency[char] = frequency.get(char, 0) + 1

print(frequency)
```
