Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis

```python
from collections import Counter
import re

search_query = "Buy mobile phone buy phone online"

# Convert to lowercase and remove punctuation
cleaned_query = re.sub(r'[.,!?;:]', '', search_query.lower())

# Split into keywords
keywords = cleaned_query.split()

# Count the frequency of each keyword
keyword_counts = Counter(keywords)

print("Keywords searched more than once:")
duplicate_keywords = {word: count for word, count in keyword_counts.items() if count > 1}
print(duplicate_keywords)
```

Problem Statement 3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour_index, reading_value in enumerate(sensor_readings):
    if reading_value % 2 == 0:  # Check if the reading is an even number
        valid_readings.append((hour_index, reading_value))

print(f"Valid Sensor Readings (Hour, Value): {valid_readings}")
```

Problem Statement 4: Email Domain Usage Analysis

```python
from collections import Counter

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domains = []
for email in emails:
    domain = email.split('@')[1]
    domains.append(domain)

domain_counts = Counter(domains)
total_emails = len(emails)

print("Email Domain Usage Analysis:")
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Define the spike threshold (30% above average)
spike_threshold = average_sales * 1.30

print(f"Average Daily Sales: {average_sales:.2f}")
print(f"Sales Spike Threshold (30% above average): {spike_threshold:.2f}")
print("Detected Sales Spikes:")

# Step 3: Detect days where sales are more than 30% above average
for day_index, sale_value in enumerate(sales):
    if sale_value > spike_threshold:
        # Day numbers are 1-based, so add 1 to the index
        print(f"Day {day_index + 1}: {sale_value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count the occurrences of each user ID
id_counts = Counter(user_ids)

print("Duplicate User ID Detection:")
# Iterate through the counts and identify duplicates
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} -> {count} times")
```
