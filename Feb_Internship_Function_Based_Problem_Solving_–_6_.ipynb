Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    # Capacity check
    if current_vehicles > capacity:
        print("Parking Status: Full / Over Capacity Alert!")
    else:
        print("Parking Status: Available")
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance_km, weather_condition, traffic_level):
    # Base delivery time (e.g., 5 minutes per km)
    base_time = distance_km * 5

    # Initialize delay
    delay = 0

    # Apply weather-based delays
    if weather_condition.lower() == "rainy":
        delay += 10  # 10 minutes delay for rain
    elif weather_condition.lower() == "snowy":
        delay += 20  # 20 minutes delay for snow
    elif weather_condition.lower() == "stormy":
        delay += 30  # 30 minutes delay for storm

    # Apply traffic-based delays
    if traffic_level.lower() == "high":
        delay += 15  # 15 minutes delay for high traffic
    elif traffic_level.lower() == "medium":
        delay += 5   # 5 minutes delay for medium traffic

    # Calculate final ETA
    estimated_time = base_time + delay

    return estimated_time

# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Calculate and display the estimated delivery time
eta = estimate_delivery_time(distance, weather, traffic)
print(f"Estimated Delivery Time: {eta} minutes")
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats_list):
    num_booked_seats = len(booked_seats_list)

    if total_seats <= 0:
        print("Error: Total seats must be greater than 0.")
        return

    occupancy_percentage = (num_booked_seats / total_seats) * 100

    print(f"Occupancy: {occupancy_percentage:.2f}%")

    show_status = ""
    suggestion = ""

    if occupancy_percentage == 100:
        show_status = "Housefull"
        suggestion = "Consider opening additional shows due to high demand!"
    elif occupancy_percentage >= 90:
        show_status = "Almost Full"
        suggestion = "Show is highly popular. Suggest opening additional shows soon."
    elif occupancy_percentage >= 70:
        show_status = "Filling Up"
    else:
        show_status = "Available"

    print(f"Show Status: {show_status}")
    if suggestion:
        print(f"Suggestion: {suggestion}")

# Sample Input 1 (from prompt)
total_seats_1 = 200
booked_seats_1 = [1] * 150  # Represents 150 booked seats

print("\n--- Analysis for Sample 1 ---")
analyze_seat_occupancy(total_seats_1, booked_seats_1)

# Sample Input 2 (Housefull example)
total_seats_2 = 100
booked_seats_2 = [1] * 100

print("\n--- Analysis for Sample 2 ---")
analyze_seat_occupancy(total_seats_2, booked_seats_2)

# Sample Input 3 (Low occupancy example)
total_seats_3 = 100
booked_seats_3 = [1] * 30

print("\n--- Analysis for Sample 3 ---")
analyze_seat_occupancy(total_seats_3, booked_seats_3)
```

 Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        print("Error: No CPU readings provided.")
        return

    average_cpu = sum(cpu_readings) / len(cpu_readings)

    server_status = ""
    if average_cpu < 50:
        server_status = "Normal"
    elif 50 <= average_cpu < 80:
        server_status = "Warning"
    else:
        server_status = "Critical"

    print(f"Average CPU Load: {average_cpu:.0f}%")
    print(f"Server Status: {server_status}")

# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

print("--- Analysis for Sample Input ---")
classify_server_load(cpu_readings)

# Additional Sample 1: Normal
print("\n--- Analysis for Normal Load ---")
classify_server_load([30, 40, 45, 25])

# Additional Sample 2: Critical
print("\n--- Analysis for Critical Load ---")
classify_server_load([85, 90, 92, 88])
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def monitor_classroom_usage(resource_usage):
    # Define overuse thresholds (example values in hours)
    overuse_thresholds = {
        "Projector": 5,  # Projector overused if used more than 5 hours
        "AC": 8,         # AC overused if used more than 8 hours
        "Lights": 6      # Lights overused if used more than 6 hours
    }

    overused_resources = []
    energy_alert = "No"

    for resource, hours_used in resource_usage.items():
        if resource in overuse_thresholds and hours_used > overuse_thresholds[resource]:
            overused_resources.append(resource)
            energy_alert = "Yes"

    if overused_resources:
        print(f"Overused Resources: {', '.join(overused_resources)}")
    else:
        print("Overused Resources: None")

    print(f"Energy Alert: {energy_alert}")

# Sample Input (from prompt)
sample_usage_1 = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

print("--- Analysis for Sample 1 ---")
monitor_classroom_usage(sample_usage_1)

# Additional Sample 2: No overuse
sample_usage_2 = {
    "Projector": 3,
    "AC": 7,
    "Lights": 5
}
print("\n--- Analysis for Sample 2 (No Overuse) ---")
monitor_classroom_usage(sample_usage_2)

# Additional Sample 3: Multiple overuse
sample_usage_3 = {
    "Projector": 7,
    "AC": 10,
    "Lights": 7
}
print("\n--- Analysis for Sample 3 (Multiple Overuse) ---")
monitor_classroom_usage(sample_usage_3)
```

 Problem Statement 6: Online Event Registration Capacity Controller


```python
def manage_event_registrations(event_capacity, registrations_list):
    num_registrations = len(registrations_list)

    if event_capacity <= 0:
        print("Error: Event capacity must be greater than 0.")
        return

    confirmed_registrations = min(num_registrations, event_capacity)
    waitlisted_users = max(0, num_registrations - event_capacity)

    registration_status = ""
    if num_registrations < event_capacity:
        registration_status = "Open"
    else:
        registration_status = "Closed"

    print(f"Confirmed Registrations: {confirmed_registrations}")
    print(f"Waitlisted Users: {waitlisted_users}")
    print(f"Registration Status: {registration_status}")

# Sample Input (from prompt)
event_capacity_1 = 100
registrations_1 = [1] * 105  # Simulates 105 registrations

print("--- Analysis for Sample 1 ---")
manage_event_registrations(event_capacity_1, registrations_1)

# Additional Sample 2: Event not full
event_capacity_2 = 50
registrations_2 = [1] * 30 # Simulates 30 registrations

print("\n--- Analysis for Sample 2 (Event Not Full) ---")
manage_event_registrations(event_capacity_2, registrations_2)

# Additional Sample 3: Event exactly full
event_capacity_3 = 75
registrations_3 = [1] * 75 # Simulates 75 registrations

print("\n--- Analysis for Sample 3 (Event Exactly Full) ---")
manage_event_registrations(event_capacity_3, registrations_3)
```
