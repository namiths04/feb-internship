***Problem 1: Social Media – Post Engagement Analyzer***


```python
# Problem 1: Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example Data
likes = [200, 300, 150, 400]

total, result = analyze_engagement(likes)

print("Total Likes:", total)
print("Post Status:", result)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
# Problem 3: Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate and round average rainfall
    average_rainfall = round(total_rainfall / len(rainfall_data))

    # Check rainfall adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


# Example
rainfall = [70, 75, 80, 65]
required = 72

average, result = check_rainfall(rainfall, required)

print("Average Rainfall:", average)
print("Rainfall Status:", result)
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
# Problem 4: Duplicate Account Detection

def detect_duplicates(usernames):

    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        return "Yes"
    else:
        return "No"


# Example
accounts = ["roshini", "kiran", "meena", "roshini"]

result = detect_duplicates(accounts)

print("Duplicate Accounts Found:", result)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(price_list):

    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    return premium


# Example
crop_prices = [1500, 2500, 1800, 3200]

premium_list = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_list)
```

**Problem 7: System Monitoring – Application Health Checker**

```python
# Problem 7: Application Health Checker

def check_system_health(error_count):

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status


# Example
errors = 7

count, result = check_system_health(errors)

print("Error Count:", count)
print("System Status:", result)
```

**Problem 8: Banking – Daily Transaction Limit Checker**

```python
# Problem 8: Daily Transaction Limit Checker

def check_transaction(amount):

    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


# Example
transaction_amount = 60000

amount, result = check_transaction(transaction_amount)

print("Transaction Amount:", amount)
print("Transaction Status:", result)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
# Problem 9: Student Attendance Eligibility

def check_attendance(attendance_list):

    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:   # 1 = Present, 0 = Absent
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Example
attendance_data = [1, 1, 0, 1, 1]

percentage, result = check_attendance(attendance_data)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", result)
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):

    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
units_consumed = 250

bill_amount, usage_status = calculate_electricity_bill(units_consumed)

print("Total Bill:", bill_amount)
print("Usage Classification:", usage_status)
```
