Problem 1: Smart Light Controller

```python
class SmartLightController:
    def __init__(self, name):
        self.name = name
        self.is_on = False  # Initial state is OFF

    def turn_on(self):
        self.is_on = True
        print(f"{self.name} is ON")

    def turn_off(self):
        self.is_on = False
        print(f"{self.name} is OFF")

    def get_status(self):
        status = "ON" if self.is_on else "OFF"
        return f"{self.name} is {status}"

# Sample Usage:
bedroom_light = SmartLightController("Bedroom Light")
print(bedroom_light.get_status())

bedroom_light.turn_on()
print(bedroom_light.get_status())

bathroom_light = SmartLightController("Bathroom Light")
bathroom_light.turn_on()
bathroom_light.turn_off()
print(bathroom_light.get_status())
```

Problem 2: Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

# Sample Usage:
employee1 = EmployeeIDCard("Rahul", "EMP102", "IT")
employee1.display_card()

print("\n---\n") # Separator for clarity

employee2 = EmployeeIDCard("Priya", "EMP105", "HR")
employee2.display_card()
```

Problem 3: Mobile Contact Record

```python
class ContactRecord:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        print("Contact Saved")

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Sample Usage:
anita_contact = ContactRecord("Anita", "9876543210")
anita_contact.display_contact()

print("\n---\n") # Separator for clarity

ravi_contact = ContactRecord("Ravi", "1122334455")
ravi_contact.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class PriceTagGenerator:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def generate_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

# Sample Usage:
headphones = PriceTagGenerator("Headphones", 2499)
headphones.generate_tag()

print("\n---\n") # Separator for clarity

laptop = PriceTagGenerator("Laptop", 65000)
laptop.generate_tag()
```

Problem 5: Movie Rating Display System

```python
class MovieRatingSystem:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")

# Sample Usage:
inception = MovieRatingSystem("Inception", 4.8)
inception.display_movie_details()

print("\n---\n") # Separator for clarity

interstellar = MovieRatingSystem("Interstellar", 4.9)
interstellar.display_movie_details()
```

Problem 6: Delivery Address Manager

```python
class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Sample Usage:
suman_delivery = DeliveryAddressManager("Suman", "Hyderabad")
suman_delivery.display_delivery_details()

print("\n---\n") # Separator for clarity

rajesh_delivery = DeliveryAddressManager("Rajesh", "Bengaluru")
rajesh_delivery.display_delivery_details()
```
